```python
import copy
import cv2
import os
import shutil
import numpy as np

# Path to the directory containing segmentation masks
# Use relative path
path = "project_agritue_dataset_2024_03_22_15_01_06_segmentation mask 1.1/SegmentationClass"
files = os.listdir(path)

# Iterate through each file in the directory
for file in files:
    name = file.split('.')[0]
    file_path = os.path.join(path, name+'.png')
    img = cv2.imread(file_path)
    H, W = img.shape[0:2]  # Get height and width of the image
    print(H, W)

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Apply Otsu's thresholding to get binary image
    ret, bin_img = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    # Find contours in the binary image
    cnt, hit = cv2.findContours(bin_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_TC89_KCOS)

    # Open a text file to write the contour coordinates
    # Use relative path
    f = open("project_agritue_dataset_2024_03_22_15_01_06_segmentation mask 1.1/SegmentationClass/{}.txt".format(file.split(".")[0]), "a+")
    for j in cnt:
        result = []
        pre = j[0]
        for i in j:
            if abs(i[0][0] - pre[0][0]) > 1 or abs(i[0][1] - pre[0][1]) > 1:  # Adjust the interval here, set to 1
                pre = i
                temp = list(i[0])
                temp[0] /= W  # Normalize x-coordinate
                temp[1] /= H  # Normalize y-coordinate
                result.append(temp)

        print(result)
        print(len(result))

        # Write contour coordinates to the text file
        if len(result) != 0:
            f.write("0 ")  # Assuming label is 0
            for line in result:
                line = str(line)[1:-2].replace(",", "")
                f.write(line + " ")
            f.write("\n")
    f.close()


```
